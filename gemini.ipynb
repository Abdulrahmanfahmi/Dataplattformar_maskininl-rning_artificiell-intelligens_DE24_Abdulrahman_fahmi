# System instructions

```python
import os
from dotenv import load_dotenv
import google.generativeai as genai


load_dotenv()
api_key = os.getenv("GOOGLE_API_KEY")
genai.configure(api_key=api_key)

system_instruction = """
You are a joking robot called Ro Båt, which will always answer with a programming joke.
"""

model = genai.GenerativeModel(
    model_name="gemini-2.5-flash",
    system_instruction=system_instruction
)

prompt = "what is the weather today?"

response = model.generate_content(prompt)

print(response.text)
```

# Temperature 

```python
model = genai.GenerativeModel("gemini-2.5-flash")

story_prompt = "write a 2 sentence story about a gray rabbit"

boring_story = model.generate_content(
    story_prompt,
    generation_config={"temperature": 0}
)

print(boring_story.text)
```

```python
model = genai.GenerativeModel("gemini-2.5-flash")

story_prompt = "write a 2 sentence story about a gray rabbit"

boring_story = model.generate_content(
    story_prompt,
    generation_config={"temperature": 2.0}
)

print(boring_story.text)
```
