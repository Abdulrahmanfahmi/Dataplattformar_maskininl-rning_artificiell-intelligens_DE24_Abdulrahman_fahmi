````python
# ALLT I EN CELL
import json
from pydantic import BaseModel
from typing import List

# 1. Definiera klasserna
class Book(BaseModel):
    title: str
    author: str
    year: int

class Library(BaseModel):
    name: str
    books: List[Book]


json_string = json.dumps(json_data)
clean_text = json_string  
library = Library.model_validate_json(clean_text)
library
````
